# Experiments

```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from ml import *
```

## Wine Quality Dataset

```python
# Preprocess dataset
wineDF = pd.read_csv("./data/winequality-red.csv")

wineDF['label'] = wineDF['quality'].apply(lambda x: 1 if x > 6 else 0)
wineDF.drop('quality', axis=1)

wine_y = wineDF['label'].values
wine_x = wineDF.drop('label', axis=1).values
```

```python
# Standardize dataset
scaler = StandardScaler()
wine_x = scaler.fit_transform(wine_x)
```

```python
# Cross validate
all_metrics = {}

for i in range(5):
    # Get the split for the data to test
    X_train, X_test, y_train, y_test = train_test_split(wine_x, wine_y, test_size=0.2, stratify=wine_y)

    logreg = LogisticRegression(C=0.046415888336127774, max_iter=100, penalty='l1',solver='liblinear')
    mlp = MLPClassifier(max_iter=200, solver='adam', hidden_layer_sizes=(50,), learning_rate_init= 0.001, activation='tanh')
    models = {'logreg': logreg, 'mlp': mlp}
    # colors = {'logreg': 'green', 'mlp': 'orange', 'decision_trees': 'blue', 'knn': 'yellow'}
    colors = {'logreg': 'green', 'mlp': 'orange'}
    decision_trees = None
    knn = None

    print(f"Split {i+1}:")
    for name, model in models.items():
        print(name)
        y_pred, y_pred_proba = fit_predict(model, X_train, X_test, y_train)
        accurracy, report_dict, report, cm, fpr, tpr, auc = get_metrics(y_test, y_pred, y_pred_proba)
        print(report_dict)
        plot_confusion_matrix(cm, plt.cm.Blues)
        all_metrics[name] = {'Accurracy': accurracy, 'Precision': report_dict['1']['precision'], 'Recall': report_dict['1']['recall'], 'F1': report_dict['1']['f1-score'], 'FPR': fpr, 'TPR': tpr, 'AUC': auc}
        # for metric, value in all_metrics[name].items():
        #     print(f" - {metric}: {value}")
    

    # Plot Other metrics
    # Plot ROC
    plt.figure()

    for model, metrics in all_metrics.items():
        plt.plot(metrics['FPR'], metrics['TPR'], linestyle='--', color=colors[model], label=model)

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves for Multiple Models')
    plt.legend()
    plt.show()

    
    
```
